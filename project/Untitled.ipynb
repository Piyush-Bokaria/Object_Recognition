```python
#object detection with distance
import cv2
import time
import supervision as sv
from ultralytics import YOLO
import pyttsx3

# Load model and set known parameters
model = YOLO("yolov8nfinal.pt")
KNOWN_WIDTH = 0.5  # Known object width in meters (adjust for specific objects)
FOCAL_LENGTH = 720  # Estimated focal length in pixels (calibrate this value)

# Class names for objects
class_names = [
    "person", "bicycle", "car", "motorcycle", "airplane", "bus", "train", "truck", 
    "boat", "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", 
    "bird", "cat", "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", 
    "giraffe", "backpack", "umbrella", "handbag", "tie", "suitcase", "frisbee", 
    "skis", "snowboard", "sports ball", "kite", "baseball bat", "baseball glove", 
    "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup", 
    "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", 
    "broccoli", "carrot", "hot dog", "pizza", "donut", "cake", "chair", "couch", 
    "potted plant", "bed", "dining table", "toilet", "tv", "laptop", "mouse", 
    "remote", "keyboard", "cell phone", "microwave", "oven", "toaster", "sink", 
    "refrigerator", "book", "clock", "vase", "scissors", "teddy bear", "hair drier", 
    "toothbrush"
]

# Initialize TTS engine
engine = pyttsx3.init()

# Initialize webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Unable to read camera feed")
    exit()

# Initialize annotators
box_annotator = sv.BoxAnnotator()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame
    frame = cv2.resize(frame, (640, 640))
    frame = cv2.GaussianBlur(frame, (5, 5), 0)

    # Object detection
    results = model.predict(frame, conf=0.25)
    detections = sv.Detections.from_ultralytics(results[0])

    # Prepare annotations for each detection
    labels = []
    speech_commands = []  # Store speech commands

    for i in range(len(detections.xyxy)):
        confidence = detections.confidence[i]
        class_id = detections.class_id[i]

        # Verify class_id is within range
        if class_id >= len(class_names):
            print(f"Warning: Detected class_id {class_id} is out of range. Skipping...")
            continue  # Skip this iteration if class_id is invalid

        # Calculate object width in pixels
        x1, y1, x2, y2 = detections.xyxy[i]
        object_width_in_pixels = x2 - x1
        
        # Ensure the width is not zero to avoid division by zero
        if object_width_in_pixels > 0:
            # Estimate distance
            distance = (KNOWN_WIDTH * FOCAL_LENGTH) / object_width_in_pixels
        else:
            distance = float('inf')  # Assign a high distance value if object width is zero

        # Create label text with object name, confidence, and distance
        label = f"{class_names[class_id]}: {confidence:.2f} - {distance:.2f}m"
        labels.append(label)  # Store label text
        speech_commands.append(f"{class_names[class_id]} at {distance:.2f} meters")  # Store speech command

    # Execute speech commands after processing all detections
    for command in speech_commands:
        engine.say(command)

    # To avoid "run loop already started" error
    if not engine._inLoop:  # Only run the loop if it's not already running
        engine.runAndWait()

    # Annotate frame with bounding boxes and labels
    annotated_image = box_annotator.annotate(scene=frame, detections=detections)

    # Draw labels on the annotated image
    for i, label in enumerate(labels):
        x1, y1, x2, y2 = detections.xyxy[i]
        cv2.putText(annotated_image, label, (int(x1), int(y1) - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    cv2.imshow("Webcam", annotated_image)

    # Delay for feedback smoothness
    time.sleep(0.5)

    # Break loop on ESC
    if cv2.waitKey(2) % 256 == 27:
        print("Escape hit, closing...")
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time
import supervision as sv
from ultralytics import YOLO
import pyttsx3  # For text-to-speech (audio feedback)

# Load your trained YOLOv8 model
model = YOLO("yolov8nfinal.pt")

# List of class names
class_names = [
    "person", "bicycle", "car", "motorcycle", "airplane", "bus", "train", "truck", 
    "boat", "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", 
    "bird", "cat", "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", 
    "giraffe", "backpack", "umbrella", "handbag", "tie", "suitcase", "frisbee", 
    "skis", "snowboard", "sports ball", "kite", "baseball bat", "baseball glove", 
    "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup", 
    "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", 
    "broccoli", "carrot", "hot dog", "pizza", "donut", "cake", "chair", "couch", 
    "potted plant", "bed", "dining table", "toilet", "tv", "laptop", "mouse", 
    "remote", "keyboard", "cell phone", "microwave", "oven", "toaster", "sink", 
    "refrigerator", "book", "clock", "vase", "scissors", "teddy bear", "hair drier", 
    "toothbrush"
]

# Initialize text-to-speech engine
engine = pyttsx3.init()

# Open the webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Unable to read camera feed")
    exit()

# Initialize annotators
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break
    
    # Preprocess the frame
    frame = cv2.resize(frame, (640, 640))
    frame = cv2.GaussianBlur(frame, (5, 5), 0)

    # Perform object detection
    results = model.predict(frame, conf=0.1)
    detections = sv.Detections.from_ultralytics(results[0]) 

    # Annotate with bounding boxes and confidence scores
    for i in range(len(detections.class_id)):
        class_id = detections.class_id[i]
        confidence = detections.confidence[i]
        class_name = class_names[class_id]
        label = f"{class_name}: {confidence:.2f}"
        box_annotator.add_label(label)

    annotated_image = box_annotator.annotate(scene=frame, detections=detections)
    
    # Display the annotated frame
    cv2.imshow('Webcam', annotated_image)

    # Text-to-Speech feedback for detected objects
    if len(detections.class_id) > 0:
        detected_objects = [f"{class_names[id]} ({detections.confidence[i]:.2f})" for i, id in enumerate(detections.class_id)]
        detected_classes_str = ", ".join(detected_objects)
        engine.say(f"Detected: {detected_classes_str}")
        engine.runAndWait()
    
    # Log performance metrics
    for i, class_id in enumerate(detections.class_id):
        print(f"Detected {class_names[class_id]} with confidence {detections.confidence[i]:.2f}")

    # Delay for smoother processing
    time.sleep(0.5)

    # Break the loop if ESC is pressed
    if cv2.waitKey(2) & 0xFF == 27:  # ESC key
        print("Escape hit, closing")
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
#writing the yaml file
import os

# Path to your label files
label_files_path = "path_to_combined_labels"  # Update with the actual path to label files
unique_class_ids = set()

# Go through each label file to gather unique class IDs
for label_file in os.listdir(label_files_path):
    if label_file.endswith(".txt"):
        with open(os.path.join(label_files_path, label_file), "r") as file:
            for line in file:
                class_id = int(line.split()[0])  # Get class ID (first element in each line)
                unique_class_ids.add(class_id)

# Sort class IDs and create placeholder names
sorted_class_ids = sorted(unique_class_ids)
class_names = [f"class_{i}" for i in sorted_class_ids]

# Save to data.yaml
with open("data.yaml", "w") as yaml_file:
    yaml_file.write("train: path/to/train/images\n")
    yaml_file.write("val: path/to/val/images\n")
    yaml_file.write("names:\n")
    for name in class_names:
        yaml_file.write(f"  - {name}\n")

print("data.yaml file created with placeholder class names.")
```
